```python
import turtle
t = turtle.Turtle()
t.pensize(2)
turtle.bgcolor('#a1def5')
t.speed(5)

# ve nha
ccao_ngoinha = 150
crong_ngoinha = 130

t.fillcolor('blue')
t.begin_fill()
t.rt(90)
t.fd(ccao_ngoinha)
t.rt(90)
t.fd(crong_ngoinha)
t.rt(90)
t.fd(ccao_ngoinha)
t.rt(90)
t.fd(crong_ngoinha)
t.end_fill()

csau_ngoinha = 100
t.fillcolor('yellow')
t.begin_fill()
t.lt(30)
t.fd(csau_ngoinha)
t.rt(120)
t.fd(ccao_ngoinha)
t.rt(60)
t.fd(csau_ngoinha)
t.rt(120)
t.fd(ccao_ngoinha)
t.end_fill()

canhmai_ngoinha = 92
t.fillcolor('#EE82EE')
t.begin_fill()
t.lt(45)
t.fd(canhmai_ngoinha)
t.lt(90)
t.fd(canhmai_ngoinha)
t.lt(135)
t.fd(crong_ngoinha)
t.end_fill()

t.fillcolor('orange')
t.begin_fill()
t.lt(30)
t.fd(csau_ngoinha)
t.lt(105)
t.fd(canhmai_ngoinha)
t.lt(75)
t.fd(csau_ngoinha)
t.lt(105)
t.fd(canhmai_ngoinha)
t.end_fill()

ccao_cuachinh = 85
crong_cuachinh = 50
t.penup()
t.goto(-40,-150)
t.pendown()
t.fillcolor('#98fb98')
t.begin_fill()
t.rt(135)
t.fd(crong_cuachinh)
t.rt(90)
t.fd(ccao_cuachinh)
t.rt(90)
t.fd(crong_cuachinh)
t.rt(90)
t.fd(ccao_cuachinh)
t.end_fill()

ccao_cuaso = 40
crong_cuaso = 35
t.penup()
t.goto(30,-25)
t.pendown()
t.fillcolor('#A52A2A')
t.begin_fill()
t.lt(120)
t.fd(crong_cuaso)
t.rt(120)
t.fd(ccao_cuaso)
t.rt(60)
t.fd(crong_cuaso)
t.rt(120)
t.fd(ccao_cuaso)
t.end_fill()


#Vẽ mặt trời
t.penup()
t.goto(200,200)
t.pendown()

tiadai = 50
tiangan = 40
bankinh_mattroi = 30
t.fd(tiadai)
t.bk(tiadai)
t.rt(45)
t.fd(tiangan)
t.bk(tiangan)
t.rt(45)
t.fd(tiadai)
t.bk(tiadai)
t.rt(45)
t.fd(tiangan)
t.bk(tiangan)
t.rt(45)
t.fd(tiadai)
t.bk(tiadai)
t.rt(45)
t.fd(tiangan)
t.bk(tiangan)
t.rt(45)
t.fd(tiadai)
t.bk(tiadai)
t.rt(45)
t.fd(tiangan)
t.bk(tiangan)
t.rt(45)

t.fillcolor('yellow')
t.begin_fill()
t.penup()
t.rt(90)
t.fd(30)
t.lt(90)
t.pendown()
t.circle(bankinh_mattroi)
t.end_fill()

# vẽ cây
t.penup()
t.goto(200,-100)
t.pendown()

# vẽ thân
ccao_than = 60
crong_than = 30
t.rt(90)
t.fillcolor('Brown')
t.begin_fill()
t.fd(crong_than)
t.rt(90)
t.fd(ccao_than)
t.rt(90)
t.fd(crong_than)
t.rt(90)
t.fd(ccao_than)
t.rt(90)
t.end_fill()

# vẽ lá
canhday_la = 120
canhben_la = 85

t.fillcolor('#98fb98')
t.begin_fill()
t.penup()
t.bk(49)
t.pendown()
t.fd(canhday_la)
t.lt(135)
t.fd(canhben_la)
t.lt(90)
t.fd(canhben_la)
t.lt(135)
t.end_fill()

t.penup()
t.lt(90)
t.fd(61)
t.rt(90)
t.pendown()
t.fillcolor('#98fb98')
t.begin_fill()
t.fd(canhday_la)
t.lt(135)
t.fd(canhben_la)
t.lt(90)
t.fd(canhben_la)
t.lt(135)
t.end_fill()

t.penup()
t.lt(90)
t.fd(61)
t.rt(90)
t.pendown()
t.fillcolor('#98fb98')
t.begin_fill()
t.fd(canhday_la)
t.lt(135)
t.fd(canhben_la)
t.lt(90)
t.fd(canhben_la)
t.lt(135)
t.end_fill()

turtle.done()
```
