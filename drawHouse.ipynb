```python
import turtle
t = turtle.Turtle()
t.pensize(2)
turtle.bgcolor('#a1def5')
t.speed(5)

# ve nha
t.fillcolor('blue')
t.begin_fill()
t.rt(90)
t.fd(150)
t.rt(90)
t.fd(130)
t.rt(90)
t.fd(150)
t.rt(90)
t.fd(130)
t.end_fill()

t.fillcolor('yellow')
t.begin_fill()
t.lt(30)
t.fd(100)
t.rt(120)
t.fd(150)
t.rt(60)
t.fd(100)
t.rt(120)
t.fd(150)
t.end_fill()

t.fillcolor('#EE82EE')
t.begin_fill()
t.lt(45)
t.fd(92)
t.lt(90)
t.fd(92)
t.lt(135)
t.fd(130)
t.end_fill()

t.fillcolor('orange')
t.begin_fill()
t.lt(30)
t.fd(100)
t.lt(105)
t.fd(92)
t.lt(75)
t.fd(100)
t.lt(105)
t.fd(92)
t.end_fill()

t.penup()
t.goto(-40,-150)
t.pendown()
t.fillcolor('#98fb98')
t.begin_fill()
t.rt(135)
t.fd(50)
t.rt(90)
t.fd(85)
t.rt(90)
t.fd(55)
t.rt(90)
t.fd(85)
t.end_fill()

t.penup()
t.goto(30,-25)
t.pendown()
t.fillcolor('#A52A2A')
t.begin_fill()
t.lt(120)
t.fd(35)
t.rt(120)
t.fd(40)
t.rt(60)
t.fd(35)
t.rt(120)
t.fd(40)
t.end_fill()


#Vẽ mặt trời
t.penup()
t.goto(200,200)
t.pendown()

t.fd(50)
t.bk(50)
t.rt(45)
t.fd(40)
t.bk(40)
t.rt(45)
t.fd(50)
t.bk(50)
t.rt(45)
t.fd(40)
t.bk(40)
t.rt(45)
t.fd(50)
t.bk(50)
t.rt(45)
t.fd(40)
t.bk(40)
t.rt(45)
t.fd(50)
t.bk(50)
t.rt(45)
t.fd(40)
t.bk(40)
t.rt(45)

t.fillcolor('yellow')
t.begin_fill()
t.penup()
t.rt(90)
t.fd(30)
t.lt(90)
t.pendown()
t.circle(30)
t.end_fill()

# vẽ cây
t.penup()
t.goto(200,-100)
t.pendown()

# vẽ thân
t.rt(90)
t.fillcolor('Brown')
t.begin_fill()
t.fd(30)
t.rt(90)
t.fd(60)
t.rt(90)
t.fd(30)
t.rt(90)
t.fd(60)
t.rt(90)
t.end_fill()

# vẽ lá
t.fillcolor('#98fb98')
t.begin_fill()
t.penup()
t.bk(49)
t.pendown()
t.fd(120)
t.lt(135)
t.fd(85)
t.lt(90)
t.fd(85)
t.lt(135)
t.end_fill()

t.penup()
t.lt(90)
t.fd(61)
t.rt(90)
t.pendown()
t.fillcolor('#98fb98')
t.begin_fill()
t.fd(120)
t.lt(135)
t.fd(85)
t.lt(90)
t.fd(85)
t.lt(135)
t.end_fill()

t.penup()
t.lt(90)
t.fd(61)
t.rt(90)
t.pendown()
t.fillcolor('#98fb98')
t.begin_fill()
t.fd(120)
t.lt(135)
t.fd(85)
t.lt(90)
t.fd(85)
t.lt(135)
t.end_fill()

turtle.done()
```
